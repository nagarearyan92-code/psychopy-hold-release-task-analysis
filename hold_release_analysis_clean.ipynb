# Hold-Release PsychoPy Task Analysis

This notebook analyses output from a simple PsychoPy hold-release task. The participant presses and holds SPACE, waits during a KEEP HOLDING period, and releases SPACE when the RELEASE NOW signal appears.

The main behavioural measure is release latency, recorded as release_response.rt.

## 1. Load packages

import pandas as pd
import matplotlib.pyplot as plt

## 2. Load PsychoPy output
The data were generated from a simple PsychoPy task in which the participant pressed and held SPACE, waited during a KEEP HOLDING period, and released SPACE when RELEASE NOW appeared.

```python
df = pd.read_csv("PSYCHOPY TASKS/errorcancdemo/data/hold_release_demo_test_output.csv")
df.head()
df.columns
display(df)

df_clean = df[["press_start.keys",
    "press_start.rt",
    "release_response.keys",
    "release_response.rt"
]].copy()
display(df_clean)
```

## 3. Select useful columns
The full PsychoPy output contains many columns. For this analysis, I keep only the key press and release timing columns needed for the behavioural analysis.

df_clean = df_clean.rename(columns={
    "press_start.keys": "press_key",
    "press_start.rt": "press_rt",
    "release_response.keys": "release_key",
    "release_response.rt": "release_rt"
})

df_clean

## 4. Add Trial Number
Rows with missing values are removed, and a trial number is added so release latency can be tracked across the task.

```python
The full PsychoPy output contains many columns. For this demo analysis, I keep only the key press and release timing columns.

df_clean["trial"] = range(1, len(df_clean) + 1)
df_clean
clean_df = df_clean.dropna()
display(clean_df)
```

## 5. Calculate mean release latency
The main measure is `release_rt`, which measures how quickly the participant released SPACE after the RELEASE NOW signal appeared.
    

```python
mean_release_rt = df_clean["release_rt"].mean()

display(mean_release_rt)
```

## 6. Plot release latency across trials
This plot shows whether release timing changed across trials, for example whether responses became faster or more variable over time.

```python
plt.plot( df_clean["trial"], df_clean["release_rt"], marker = "o")
plt.xlabel("Trial")
plt.ylabel("Release RT (seconds)")
plt.title("Release RT across hold-release trials")
plt.show()
```

## 8. Brief interpretation

This demo confirms that PsychoPy can record response-release timing using the Keyboard component set to register key release. The key variable is `release_rt`, which measures how quickly the participant released SPACE after the RELEASE NOW signal appeared.

In this test run, release latency varied across trials, and the mean release RT provides a simple behavioural measure of release speed. This workflow is relevant for future dissertation work because release timing may be used as a behavioural marker of error awareness in a press-and-hold task.
